```python
from langchain_community.embeddings import OllamaEmbeddings
embeds = OllamaEmbeddings(model="gemma2:2b")
```

```python
texts = [
    'Alpha is the first character of the Greek alphabet.',
    'Beta is the second letter of the Greek alphabet.'
]
result = embeds.embed_documents(texts)
```

This is the embedding of the First text "Alpha is the first character of the Greek alphabet."

```python
# dimensions of gemma:2b
len(result[0])
```

```python
result[0]
```

## Embedding a Query
To embed a query, you can use the same embedding method with a single sentence.

```python
query = embeds.embed_documents("what is the second letter of Greek alphabet")
query
```
